```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsRegressor, KNeighborsClassifier
from sklearn.metrics import accuracy_score, balanced_accuracy_score
```

```python
data = pd.read_csv('glass.csv')
#https://www.kaggle.com/datasets/uciml/glass
data
```

```python
mms = MinMaxScaler()
data['RI'] = mms.fit_transform(data[['RI']])
data['Na'] = mms.fit_transform(data[['Na']])
data['Mg'] = mms.fit_transform(data[['Mg']])
data['Al'] = mms.fit_transform(data[['Al']])
data['Si'] = mms.fit_transform(data[['Si']])
data['K'] = mms.fit_transform(data[['K']])
data['Ca'] = mms.fit_transform(data[['Ca']])
data['Ba'] = mms.fit_transform(data[['Ba']])
data['Fe'] = mms.fit_transform(data[['Fe']])

#data = data.drop(columns=['RI'])
#data = data.drop(columns=['Na'])
#data = data.drop(columns=['Ba'])
#data = data.drop(columns=['Si'])
#data = data.drop(columns=['K'])
#data = data.drop(columns=['Ca'])
#data = data.drop(columns=['Fe'])

#data = data.drop(columns=['Fe'])
data.head()
```

```python
#sns.heatmap(data.corr(), annot=True, fmt='.3f')

def make_cot(data):
    data['cotType'] = data['Type'].map({1: 'A',2:'B', 3:'C', 4: 'D', 5:'E',6:'F',7:'G'})
    
```

```python
X_data, y_data = data.drop(columns=['Type']), data[['Type']]
X_test, X_train, y_test, y_train = train_test_split(X_data, y_data,test_size=0.2, random_state=1)
#X_train['Type'] = y_train
test_data = X_test
test_data['Type'] = y_test['Type']
Train_data = X_train
Train_data['Type'] = y_train['Type']
```

```python
make_cot(y_train)
make_cot(y_test)
X_test = X_test.drop(columns=['Type'])
X_train = X_train.drop(columns=['Type'])
```

```python
KNeighborsRegressorObj.fit(X_test, y_test['Type'])
KNeighborsRegressorObj.predict(X_train)
```
